# Investigating differences
Next, we will create two histograms that show the distribution in values for 'Area (mean)' for both benign and malignant tumors.  
To permit easy comparison, we created a single figure containing both histograms in the code cell below._)

```
sns.distplot(a=cancer_b_data['Area (mean)'], label='Benign', kde=False)
sns.distplot(a=cancer_m_data['Area (mean)'], label='Malignant', kde=False)
plt.title('Distribution in values for Area (mean)')
plt.legend()
```

```
sns.kdeplot(data=cancer_b_data['Area (worst)'], shade=True, label='Benign',)
sns.kdeplot(data=cancer_m_data['Area (worst)'], shade=True, label='Malignant')
```

##Joint KDEplot:

```
sns.jointplot(x=cancer_b_data['Area (worst)'],y=cancer_b_data['Area (worst)'], kind= 'kde')
```
